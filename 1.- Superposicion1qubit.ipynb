```python
# UN QUBIT CON UNA PUERTA H Y SU MEDIDA

# Importar las librerías necesarias
from qiskit import QuantumCircuit  
# QuantumCircuit es una clase que permite crear circuitos cuánticos.
from qiskit_aer import Aer      
# Aer es un simulador de Qiskit que permite ejecutar circuitos cuánticos en un entorno simulado.
import matplotlib.pyplot as plt
# matplotlib.pyplot es una biblioteca de Python para crear gráficos y visualizaciones.


qc_a = QuantumCircuit(1, 1)  # Definimos un circuito de 1 qubit y un bit
qc_a.h(0)                    # Añadimos un apuerta Hadamard

qc_a.measure(0,0)            # Medimos el qubit 0 en el bit 0
qc_a.draw("mpl")             # Dibujamos el circuito cuántico
# mpl es un backend de matplotlib que permite visualizar el circuito cuántico.
```

```python
#  LO EJECUTAMOS 10.000 VECES Y VEMOS LOS RESULTADOS

# Configuración del simulador cuántico
Procesando = Aer.get_backend('qasm_simulator')        # Obtiene el backend 'qasm_simulator' de Qiskit Aer
#qasm_simulator es un simulador cuántico que permite ejecutar circuitos cuánticos y obtener resultados de medidas.

# Ejecución del circuito cuántico en el simulador cuántico
Proceso = Procesando.run(qc_a, shots=10000).result()  # Ejecuta el circuito cuántico 10000 veces y obtiene los resultados
# shots=10000 indica que el circuito se ejecutará 10.000 veces para obtener una distribución estadística de los resultados.
# qc_a es el circuito cuántico que se va a ejecutar.
# run() ejecuta el circuito cuántico en el backend especificado.
# result() devuelve un objeto que contiene los resultados de la ejecución del circuito cuántico.

Resultados = Proceso.get_counts(qc_a)                 # Obtiene los conteos de los resultados de la ejecución del circuito
print(Resultados)                                     # Imprime los resultados en la consola
```

```python
# Generar el histograma para ambas columnas
#
# Extraer etiquetas (eje x) y valores (eje y)
etiquetas = list(Resultados.keys())# obtiene 0 y1   
valores = list(Resultados.values()) # obtiene las veces que se repide el estado 0 y 1

# Crear gráfico de barras
plt.bar(etiquetas, valores, color=['skyblue', 'lightcoral'])

# Agregar el valor encima de cada barra
for i, valor in enumerate(valores):
    plt.text(i, valor + max(valores)*0.01, str(valor), ha='center', va='bottom', fontsize=12)

# Etiquetas y título
plt.xlabel('Categorías')
plt.ylabel('Frecuencia')
plt.title('Histograma de Resultados')

# Mostrar gráfico
plt.tight_layout()
plt.show()
```
